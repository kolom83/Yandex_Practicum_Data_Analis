Lambda-функции \
Короткая запись функции, у которой не своего названия и подразумевается ее использование только один раз
```
lambda x: x/sum(x) =>

def _(x):
    return x/sum(x)
```



__map__ можно использовать только для Series\
Функция, которая является параметром в map применяется к каждому элементу Series по отдельности

````python
# вызов функции
def makeNames(df):
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    df.columns.str.contains('nearest') #присоединяет префикс
    df = df.add_suffix('_km') # прификсация наименований колонок
    df['poselok'] = df.locality_name.str.find('поселок') >= 0
    
    df['area_type'] = df.locality_name.fillna('').apply(
        lambda title: ' '.join(word for word in title.split(' ') if word.islower()) # разбиение строки на список значений
    
    df['id'] = pd.to_numeric(df['id'], errors='coerce') #используется для преобразования переданного аргумента в числовой тип
    df['dt'] = pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M') #преобразует даты, представленные в разных форматах, в объекты DateTime format='%Y-%m-%dT%H:%M:%S'
    pd.to_timedelta() преобразование в тип Timedelta    
    
    df['week'] = df['date'].dt.isocalendar().week
    # формирование столбца с номером месяца в формате дата
    df['month'] = df['dt'].values.astype('datetime64[M]')
    df['month'] = df['date'].dt.isocalendar().month
    # Приведение к типу datetime64[M]
    df['month'].dt.to_period('M')
    df['month'] = pd.Grouper(freq='M').df['timestamp']
    df['month'] = pd.DatetimeIndex(df['timestamp']).month
   
    # проверка на вхождения значения из одного столбца в другой для строковых и для численных типов данных
    df['payer'] = df['user_id'].str.contains(df['user_id']).any()
    # isin - это метод, который позволяет фильтровать данные в DataFrame или Series на основе проверки, 
    # содержится ли значение в определенном списке значений.    
    df['payer'] = df['user_id'].isin(df['user_id'])    
     
    # добавляем колонку, в которой значения NAN будут как 0 а не NAN как 1
    df['active'] = ~df['revenue'].isna()*1   
        
    df['a'] = df['a'].astype('category') #перевоидит тим в категориальный тип, экономит место
    
    df['group_name'] = df['name'].where(df['name'].isin(big_nets_stat.index), 'Другие') #имеет значение True, сохраняется исходное значение    
        
    return df

# основной блок
(
    df
    .drop(columns=['key'],axis=1,inplace=True) # удаление колонки
    .droplevel(['payer', 'channel']) # устраните payer и channel из индексов 
    .set_index('dt') # в индекс датафрейма передается колонка с датой
    .isna() #ищет пропуски
    .isnull()
    .notnull()
    .fillna(value={'city': df.groupby(['client_id'])['city'].transform('last')}, #замена нуля последним значением в сгруппированных данных
            method = {'bfill', 'ffill', None}, # можно брать соседнии значения
            axis = {0, 1}) # по строкам или колонкам
    .drop_duplicates(subset='client_id', keep='last') #дубликаты по колонке, удаляем все значения кроме последнего 'last' или  последней 'first'
    .pipe(makeNames) #вызов функции для собственной обработки
    .assign(#создание новых колонок
        fancy_column=lambda df: df["sepal_width"]["mean"]/ df["sepal_width"]["mean"].mean(),
        useless_column="I am useless",
        to_big_to_small=lambda df: (df.sepal_width > 3).map({True: "Too Big", False: "Perfect"}), # выделение категорий
        inverted_name=lambda df: df.species.map(lambda name: name[::-1]), #С помощью map можно заменить любое значение в серии другим значением (например строкой на оборот) 
        R = cut(df['recency'], bins=3, labels=['3', '2', '1']), # разбиение на интервалы 
        credit_group2 = pd.cut(df['credit_amount'],[0,1000,2000,3000,5000,np.inf]) # или так разбиение на интервалы
        F_qcut = pd.qcut(data['frequency'], q=3, labels=['1', '2', '3']) # разбиение на квантили
        median_price = df.groupby('locality_name')['last_price'].transform(np.median) # получения медианы по группировкам для основного датфрейма
    ) 
    .groupby('region', #группировка
            as_index=False) #переиндексирует
    .agg(  #агригирующие процдуры
            user_count = ('user_id', 'count'), # колоичество     
            user_nunique = ('user_id', 'nunique'), # колоичество различных
            user_payer = ('payer', 'sum'),
            user_payer_mean = ('payer', 'mean'),
            user_payer_median = ('payer', 'median'),
            user_payer_max = ('payer', 'max'), #min
            user_payer_last = ('payer', 'first'), #last
    
        ).idxmin() или .idxmax() .size() возвращает размеры групп (возвращает Series)
    .rename(columns={'Uid': 'uid', 'Device':'device', 'End Ts':'end_ts', 'Source Id':'source'}) #переименоваине строк
    .query("PW Squared > 0.4 and petal_length > @length_th and species != 'setosa' or поле in  not in ['S7', 'Rossiya']") #запрос
    .explode("city_to_list") # развернуть списки внутри одной таблицы, увеличивает колоичество строк
    .reset_index() #переиндексация
    .sort_values(by=['user_count','date'], ascending=False) #сортировка
    .astype({"user_id": str, "order_id": str}) #изменение типа
    .to_frame() #выводит в датафрейм
    .replace({ # категорирование
                "status_int": {"Заказ зарезервирован": 1, "Заказ упакован": 2, "Заказ отгружен со склада": 3}
             })
)
````
