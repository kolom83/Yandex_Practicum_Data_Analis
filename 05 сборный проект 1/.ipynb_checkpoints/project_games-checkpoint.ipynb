**Описание проекта**

Интернет-магазине «Стримчик» продаёт по всему миру компьютерные игры. Доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы. Нужно выявить определяющие успешность игры закономерности. 

Данные до 2016 года. Необходимо спланировать кампанию на 2017-й. 

В наборе данных есть рейтинг ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

# Шаг 5. Проверьте гипотезы

```python
user_score_XOne = df.query('platform == "XOne"').user_score.dropna()
user_score_XOne.mean()
```

```python
user_score_PC = df.query('platform == "PC"').user_score.dropna()
user_score_PC.mean()
```

Нулевая гипотеза, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, альтернативная гипотеза двухсторонняя. Две независимые выборки, метод «scipy.stats.ttest_ind». 

```python
alpha = 0.05# задайте уровень стат. значимости

results = st.ttest_ind(user_score_XOne, user_score_PC,
                        # alternative='greater',
                        # alternative='less'
                      )

# выведите на экран получившееся p-значение
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу о равенстве средней пользовательской оценки двух платформ')
else:
    print('Не получилось отвергнуть нулевую гипотезу о равенстве средней пользовательской оценки двух платформ')
```

Средний рейтинг этих платформ не одинаков.  

```python
user_score_Action = df.query('genre == "Action"').user_score.dropna()
user_score_Action.mean()
```

```python
user_score_Sports = df.query('genre == "Sports"').user_score.dropna()
user_score_Sports.mean()
```

Проверим гипотезу, средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза: средние пользовательские рейтинги жанров одинаковы.

Альтернативная гипотеза средний пользовательский рейтинг жанра Action больше рейтинга жанра Sports.


```python
alpha = 0.05# задайте уровень стат. значимости

results = st.ttest_ind(user_score_Action, user_score_Sports, 
                         alternative='greater',
                         # alternative='less'
                      )

# выведите на экран получившееся p-значение
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Результат: нулевая гипотеза отвергнута в пользу альтернативной гипотезы: средний пользовательский рейтинг жанра Action больше рейтинга жанра Sports.

        o	удалил две строки с пустыми именем и жанром; 
        o	пропусков в оценках сильно много (около 8 тысяч строк), замена их нулем, средним и медианой не имеет смысла даже в рамках платформ;

        o	Пустые значения в оценках может быть связана с отсутствием таких оценок. Пропуски в столбце с годами более важны, на это надо обратить внимание.

_Шаг 3. Исследовательский анализ данных_

    •	Больше всего было выпущено и продано игр в 2008 и 2009 году. До 1990 года выпускалось мало игр. Данными до 1990 года можно пренебречь. Больше всего играют в Северной Америке и в Европе. 
    •	Чем больше было выпущено игр для платформы, тем больше были продижи. Лидерами за все время являются PlayStation, Xbox, Nintendo, Wii. В среднем версия игровой платформы живет около семи лет. Для прогноза на 2017 год достаточно взять данные за предыдущие шесть лет, с 2010 года. 
    •	Данные с 2010 года составляют около трети всех данных.
    •	За последние семь лет наметилась общая тенденция к снижению продаж игр на всех игровых платформах. 
    •	Лидерами продаж к 2016 году являются PS4, XOne, 3DS. И хотя наибольшие продажи за последние 7 лет были у PS3, X360, Wii, время этих платформ проходит и вряд ли они будут лидерами продаж в будущем.   
    •	Продажи практически не коррелируются с оценкой пользователей и положительно коррелируются с оценкой критиков с коэффициентом около 0,4. Важнее для продаж оценка профессиональных экспертов, чем пользователей. Возможно при начале продаж оценок пользователей еще мало, и покупатели ориентируются больше на экспертные оценки.
    •	Самые популярные и продаваемые игры, это экшен, спорт, ролевые игры  (ходилки, бролилки, стрелялки от первого лица), не самые продаваемые - головоломки, стратегии. Напрягать мозги лишний раз желающих мало, а вот побегать, хоть виртуально, много.

_Шаг 5. Проверка гипотез_

    •	Нулевая гипотеза, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, альтернативная гипотеза двухсторонняя. Две независимые выборки, метод «scipy.stats.ttest_ind». Результат: нулевая гипотеза о равенстве средней пользовательской оценки двух платформ отвергнута. Средний рейтинг этих платформ не одинаков.
    •	Проверим гипотезу, средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные. Нулевая гипотеза: средние пользовательские рейтинги жанров одинаковы. Альтернативная гипотеза: средний пользовательский рейтинг жанра Action больше рейтинга жанра Sports. Результат: нулевая гипотеза отвергнута в пользу альтернативной гипотезы: средний пользовательский рейтинг жанра Action больше рейтинга жанра Sports.
    
_Шаг 6. Общий вывод_

	Представлены данные о выпусках и продажах компьютерных игр по годам с данными о продажах, оценках критиков и пользователей, возрастном рейтинге. Около трехсот строк не имеют года выпуска, что важно для исследования, поэтому эти строки отброшены. Около половины записей не имею данные об оценках и рейтинге, что не сильно критично, заполнять их агрегированными данными не имеет смысла, оставляем как есть. 
    
	Исследование показало, что основной выпуск и продажи игр пришлись на нулевые и десятые годы. Пикам стали 2008-2009 года, с тех пор и выпуск, и продажи снижаются. Продолжительность активных продаж в рамках одной платформы ограниченно 5-7 годами. Актуальный период для прогнозирования продаж на 2017 год является период 2010 по 2016. За этот период лидерами продаж по объему были PS3, X360, Wii, но их популярность проходит. Наиболее перспективными являются PS4, XOne, 3DS.
    
	Была выявлена корреляция между экспертными оценками и продажами. Зависимости продаж от пользовательских оценок нет. Наиболее популярными жанрами являются экшен-игры, спортивные симуляторы, шутеры. Есть региональные особенности, в Японии так же популярны ролевые игры. Есть региональные различия и в платформах. Игры со взрослым рейтингом популярны в Северной Америки, а игры с подростковым рейтингом или рейтингом для всех в Японии.
    
    Гипотеза о равенстве пользовательского рейтинги платформ Xbox One и PC не подтвердилась.
    Гипотеза о превышении пользовательского рейтинги жанра Action над  жанром Sports не была опровергнута.
