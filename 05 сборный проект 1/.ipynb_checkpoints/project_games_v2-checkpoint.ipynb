```python
# –∏—Å–ø–æ–ª—å–∑—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫—É Plotly
# —Å–æ–∑–¥–∞–µ–º –ª–∏–Ω–∏—é –¥–ª—è —á–∏—Å–ª–∞ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π
trace0 = go.Scatter(
    x=years_df.index,
    y=years_df.world_sales,
    name='–ú–∏—Ä–æ–≤—ã–µ –ø—Ä–æ–¥–∞–∂–∏'
)

# —Å–æ–∑–¥–∞–µ–º –ª–∏–Ω–∏—é –¥–ª—è —á–∏—Å–ª–∞ –≤—ã—à–µ–¥—à–∏—Ö –∏–≥—Ä 
trace1 = go.Scatter(
    x=years_df.index,
    y=years_df.number_of_games,
    name='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–ø—É—â–µ–Ω–Ω—ã—Ö –∏–≥—Ä'
)

# –æ–ø—Ä–µ–¥–µ–ª—è–µ–º –º–∞—Å—Å–∏–≤ –¥–∞–Ω–Ω—ã—Ö –∏ –∑–∞–¥–∞–µ–º title –≥—Ä–∞—Ñ–∏–∫–∞ –≤ layout
data = [trace0, trace1]
layout = {'title': '–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –≤–∏–¥–µ–æ –∏–≥—Ä'}

# c–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç Figure –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –µ–≥–æ
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
# —É–¥–∞–ª–∏–º –∑–∞–ø–∏—Å–∏ –¥–æ 1995 –≥–æ–¥–∞
df = df.query('year_of_release >= 1995')
```

__–ü–æ—Å–º–æ—Ç—Ä–∏—Ç–µ, –∫–∞–∫ –º–µ–Ω—è–ª–∏—Å—å –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º.__

```python
platforms_df = (
    df.groupby("platform")[["world_sales"]]
    .sum()
    .join(df.groupby("platform")[["name"]].count())
)
platforms_df.columns = ["world_sales", "number_of_games"]
platforms_df.sort_values("world_sales", ascending=False, inplace=True)
# platforms_df
```

```python
trace0 = go.Bar(
    x=platforms_df.index, 
    y=platforms_df.world_sales, 
    name="–ú–∏—Ä–æ–≤—ã–µ –ø—Ä–æ–¥–∞–∂–∏")

trace1 = go.Bar(
    x=platforms_df.index,
    y=platforms_df.number_of_games,
    name="–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–ø—É—â–µ–Ω–Ω—ã—Ö –∏–≥—Ä",
)

data = [trace0, trace1]
layout = {"title": "–î–æ–ª—è –ø–ª–∞—Ç—Ñ–æ—Ä–º"}

fig = go.Figure(data=data, layout=layout)

iplot(fig, show_link=False)
```

```python
# –≤—ã–±–µ—Ä–µ–º —Ç–æ–ø 10 –Ω–∞–∏–±–æ–ª–µ–µ –ø—Ä–∏–±—ã–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏ –æ—Ç–±–µ—Ä–µ–º –ø–æ –Ω–∏–º –¥–∞–Ω–Ω—ã–µ –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã
top_platform = df.groupby('platform')['world_sales'].sum().sort_values(ascending=False).nlargest(5).index.values
df_top_platform = df.query('platform in @top_platform')
```

```python
# —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–ª–∏—Ü—É, –≥–¥–µ –ø–æ —Å—Ç—Ä–æ–∫–∞–º –≥–æ–¥–∞, –ø–æ –∫–æ–ª–æ–Ω–∫–∞–º –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã, –≤ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–∏ —Å—É–º–º—ã –ø—Ä–æ–¥–∞–∂ –ø–æ –º–∏—Ä—É
df_top_platform_sales = df_top_platform.pivot_table(index='year_of_release', columns = 'platform', values='world_sales', aggfunc='sum')
```
