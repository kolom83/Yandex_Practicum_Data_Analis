```python
def title_parse(title):
    res = ' '.join(word for word in title.split(' ') if word.islower())
    return res
```

```python
def title_parse_(title):
    res = ' '.join(word for word in title.split(' ') if word.islower()==False)
    return res
```

```python
# print( title_parse_('Санкт-Петербург'))
# в колонке 'locality_name' оставим только имена собственные на заглавную букву
data['locality_name'] = data.locality_name.apply(title_parse_)
```

```python
def chek_floors(row):
    result = row['floors_total']
    if result < row['floor']:
        result = row['floor']
    
    return result
```

```python
data['floors_total'] = data.apply(chek_floors, axis=1)
```

Пропуски в расстоянии до центра города могут быть связаны с тем, что объект находится не в черте города, и при размещении объявления это значение просто не указывают. Заменить его на среднее или медиальное или модальное значение может дать ошибку.   

Пустые значения в колонках расстояние до парка, количество парков во круг или тоже самое о водоемах, может говорить просто об отсутствии подобных объектов. Замещать пустые значения какими ни будь агрегированными данными не имеет смысла, можно добавить колонку, в которой внести булевское значение наличия паков или водоемов в округе.

```python

def floor_type_(row):
    floor = row['floor']
    floors_total = row['floors_total']
    try:
        if floor == 1:
            return 'первый'
        elif floor == floors_total:
            return 'последний'
        else:
            return 'другой'
    except:
        pass
```
