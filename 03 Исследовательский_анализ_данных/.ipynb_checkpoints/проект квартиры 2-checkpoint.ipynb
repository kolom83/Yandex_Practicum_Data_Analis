```python

def floor_type_(row):
    floor = row['floor']
    floors_total = row['floors_total']
    try:
        if floor == 1:
            return '–ø–µ—Ä–≤—ã–π'
        elif floor == floors_total:
            return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
        else:
            return '–¥—Ä—É–≥–æ–π'
    except:
        pass
```
