В вашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Пропущенные значения в колонке балконы заменяем на 0, если не указано то и балконов нет.

```python
data['balcony'] = data['balcony'].fillna(0)
```

Если значение в колонке апартаменты пуста, значит это не апартаменты.
При переводе в тип булева значения NAN станут False.

```python
data['is_apartment'] = data['is_apartment'].astype('bool')
```

Пропущенные значения в колонки локация заменим на "не указано".

```python
data['locality_name'] = data['locality_name'].fillna('не указано')
```

Ищем и правим неявные дубли в названиях локаций. 

```python
data['locality_name'] = (
    data['locality_name']
    
    .str.replace('ё', 'е')
    .replace(['городской поселок', 'поселок городского типа', 'коттеджный поселок'], 'поселок', regex=True)
) 
```

Проверяем уникальные значения locality_name.

Потолки

```python
data['ceiling_height'].describe()
```

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
```

Доля пропусков 38%. Аномально высоких и низких потолков мало, стандартное отклонение не сильное, по этому пустые значения можно заполнить медианным значением. 

floors_total — всего этажей в доме

```python
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].mode())
# data['floors_total'].fillna(data['floors_total'].mode(), inplace=True)
```

```python
def chek_floors(row):
    result = row['floors_total']
    if result < row['floor']:
        result = row['floor']
    
    return result
```

```python
data['floors_total'] = data.apply(chek_floors, axis=1)
```

Пропущенных значений в колонке всего этажей не много, можно заменить чаще всего встречаемым значением, а потом если этаж оказался выше этажности, привести этажность к значению указанного этажа.  

airports_nearest — расстояние до ближайшего аэропорта в метрах (м)

```python
data['airports_nearest'].describe()
```

```python
data.pivot_table(index = data.index, values='airports_nearest').hist(bins=100)
plt.show()
```

```python
data['airports_nearest'] = data['airports_nearest'].fillna(data['airports_nearest'].mean())
```

Распределение значений расстояния до аэропорта распределено относительно нормально, можно использовать для пустых значений среднее. 

cityCenters_nearest — расстояние до центра города (м)

```python
data.pivot_table(index = data.index, values='cityCenters_nearest').hist(bins=100)
plt.show()
```

Пропуски в расстоянии до центра города могут быть связаны с тем, что объект находится не в черте города, и при размещении объявления это значение просто не указывают. Заменить его на среднее или медиальное или модальное значение может дать ошибку.   

Пустые значения в колонках расстояние до парка, количество парков во круг или тоже самое о водоемах, может говорить просто об отсутствии подобных объектов. Замещать пустые значения какими ни будь агрегированными данными не имеет смысла, можно добавить колонку, в которой внести булевское значение наличия паков или водоемов в округе.

### Добавьте в таблицу новые столбцы

Добавим стоимость метра квадратного. 

```python
data['price_square_meter'] = round(data['last_price']/data['total_area'], 2)
```

```python
# добавим дни недели
data['weekday'] = pd.DatetimeIndex(pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')).weekday
```

```python
# добавим месяц
data['month_publication'] = pd.DatetimeIndex(pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')).month
```

```python
# добавим год
data['year_publication'] = pd.DatetimeIndex(pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')).year
```

```python

def floor_type_(row):
    floor = row['floor']
    floors_total = row['floors_total']
    try:
        if floor == 1:
            return 'первый'
        elif floor == floors_total:
            return 'последний'
        else:
            return 'другой'
    except:
        pass
```

```python
# определяем этажность квартир
data['floor_type'] = data.apply(floor_type_, axis=1)
```

```python
data['floor_type'].value_counts()
```

```python
# перевод расстояния до центра города в километры
data['cityCenters_nearest'] = round((data['cityCenters_nearest']/1000))
```

Пропуски в расстоянии до центра города могут быть связаны с тем, что объект находится не в черте города, и при размещении объявления это значение просто не указывают. Заменить его на среднее или медиальное или модальное значение может дать ошибку. 

Были добавлены данные стоимости квадратного метра, днях недели, месяцах и годах размещения объявлений. 
При анализе были выявлены некорректные данные и выбросы о высотах квартир. 
Для количественных данных были построены коробки с усами, рассчитаны пороги для оценок выбросов. 

При удалении выбросов количество данных сократилось на 20 процентов. 

Анализ времени продаж показал: время продаж имеет распределение Пуассона. 
К быстрым продажам можно отнести продажи с временем ниже медианного значения (95 дней), 
к медленным – продажи со временем выше 3 перцентили (более 200 дней).       

Корреляция показателей сводной таблицы по ценам и разным показателям выявило зависимость цены от общей площади 
и слабую зависимость цены от жилой площади и площади кухни. Так же квартиры на первом и последнем дешевле других. 

Цена от дня недели или месяца года не зависит, может только зимой цены чуть выше. 
При этом цены с годами растут, а высокая стоимость в первый наблюдаемый год объясняется маленькой выборкой. 

Больше всего и самые дорогие объявления по Санкт-Петербургу. 
Самая высокая средняя цена в Кронштате (может курортная зона), самая маленькая средняя цена в деревне Старополье. 
Чем дальше от центра Санкт-Петербурга, тем цена квадратного метра дешевле.
